```python

import cv2
from PIL import Image
from IPython.display import display
```

```python
image = cv2.imread('./Downloads/img.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

## Convert image to a gray scale image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

## Convert to invert grayscale image

```python
invert_image = cv2.bitwise_not(gray_image)
display(Image.fromarray(invert_image))
```

## Negative image to blurry image

```python
blurred_image = cv2.GaussianBlur(invert_image, (21,21), 0)
display(Image.fromarray(blurred_image))
```

## Pencil Sketch image with the inverted blurry image

```python
inverted_blurred=cv2.bitwise_not(blurred_image)
pencil_sketch = cv2.divide(gray_image, invert_image, scale=250.0)
display(Image.fromarray(pencil_sketch))
```

### Other codes

```python
import cv2
import matplotlib.pyplot as plt

# Load the PNG image
image = cv2.imread('./Downloads/l5.jpg')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Invert the grayscale image
inverted_image = cv2.bitwise_not(gray_image)

# Blur the inverted image
blurred_image = cv2.GaussianBlur(inverted_image, (3, 3), 0)

# Divide the grayscale image by the blurred image
sketch_image = cv2.divide(gray_image, blurred_image, scale=255.0)

# Display the original image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Display the sketch image
plt.subplot(1, 2, 2)
plt.title('Sketch Image')
plt.imshow(sketch_image, cmap='gray')
plt.axis('off')

# Show the plots
plt.show()
```

```python

# Load the PNG image
image = cv2.imread('./Downloads/l.jpg')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Invert the grayscale image
inverted_image = cv2.bitwise_not(gray_image)

# Blur the inverted image
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), sigmaX=0, sigmaY=0)

# Divide the grayscale image by the blurred image
sketch_image = cv2.divide(gray_image, blurred_image, scale=230.0)

# Set a larger figure size
fig = plt.figure(figsize=(20, 20))

# Display the original image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Display the grayscale image
plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Display the sketch image
plt.subplot(1, 3, 3)
plt.imshow(sketch_image, cmap='gray')
plt.title('Sketch Image')
plt.axis('off')

# Show the plots
plt.show()
```
